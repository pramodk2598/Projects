#Correlation

```python
cat_df = df[cat_cols].apply(lambda x : pd.factorize(x)[0])
cat_df.info()
```

```python
concat_df = pd.concat([cat_df, df[int64_cols]], axis=1)
concat_df.info()
```

```python
sns.heatmap(concat_df.corr(), cmap='Oranges')
```

Examining Attrition_Age with other feature doesn’t show us any correlation. However, by using Chi-square test we can see that some features are dependent with the target. I will use chi2_contingency to compute chi2 and to get p-values. p-value is the Probability of H0 being True. if p-Value>0.05 then Accept the assumption(H0). The assumption H0 is our case will be independence.

```python
from scipy.stats import chi2_contingency
factors_paired = [(i,j) for i in concat_df.columns.values for j in concat_df.columns.values] 

chi2, p_values =[], []

for f in factors_paired:
    if f[0] != f[1]:
        chitest = chi2_contingency(pd.crosstab(concat_df[f[0]], concat_df[f[1]]))   
        chi2.append(chitest[0])
        p_values.append(chitest[1])
    else:      # for same factor pair
        chi2.append(0)
        p_values.append(0)
n_features = len(concat_df.columns)
chi2 = np.array(chi2).reshape((n_features, n_features)) # shape it as a matrix
chi2 = pd.DataFrame(chi2, index=concat_df.columns.values, columns=concat_df.columns.values) # then a df for convenience
# p values is what will help us to decide dependency or independency
p_values = np.array(p_values).reshape((n_features, n_features))
p_values = pd.DataFrame(p_values, index=concat_df.columns.values, columns=concat_df.columns.values)
```

```python
column_to_drop = ['Attrition_Flag',
                  'Card_Category',
                  'Dependent_count',
                  'Customer_Age_period',
                  'Months_on_book_Period',
                  'Marital_Status']

column_to_drop_2nd_try = ['Attrition_Flag',
                          'Card_Category',
                          'Dependent_count',
                          'Customer_Age',
                          'Months_on_book',
                          'Marital_Status']

ordinal_cols = ['Income_Category', 'Education_Level']
nominal_cols = ['Gender']
cols_to_scale = ['Customer_Age',
                'Months_on_book',
                'Total_Revolving_Bal',
                'Total_Trans_Amt',
                'Total_Trans_Ct',
                'Credit_Limit',
                'Avg_Open_To_Buy']
```

```python
from sklearn import set_config
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OrdinalEncoder, OneHotEncoder
from sklearn.compose import make_column_transformer, ColumnTransformer
from sklearn.model_selection import train_test_split

set_config(display='diagram')
```

```python
imp_enc = SimpleImputer(missing_values='Unknown', strategy='most_frequent')

ord_enc = OrdinalEncoder(categories=[
                ['Less than $40K', '$40K - $60K', '$60K - $80K', '$80K - $120K', '$120K +'],
                ['Uneducated', 'High School', 'College', 'Graduate', 'Post-Graduate', 'Doctorate']])

one_hot_enc = OneHotEncoder(handle_unknown='error', drop='if_binary')

target_enc = OneHotEncoder(categories=[['Attrited Customer', 'Existing Customer']], 
                           handle_unknown='error', 
                           drop='if_binary',
                           sparse=False)

ordinal_transformer = Pipeline(steps=[('unknown_imp', imp_enc), ('ordinal', ord_enc)])

numeric_transformer = Pipeline(steps=[('num_imp', SimpleImputer(strategy='median')),
                                       ('scaler', StandardScaler())])

nominal_transformer = Pipeline(steps=[('unknown_imp', imp_enc), ('one_hot', one_hot_enc)])

preprocessor = make_column_transformer((ordinal_transformer, ordinal_cols),
                                       (nominal_transformer, nominal_cols),
                                       (numeric_transformer, cols_to_scale),
                                        remainder = 'passthrough') 
preprocessor
```

```python
df_to_fit = df.drop(columns=column_to_drop )
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_to_fit,
                                                    target_enc.fit_transform(df[['Attrition_Flag']]).ravel(),
                                                    test_size=0.3,
                                                    random_state=42, 
                                                    stratify=target_enc.fit_transform(df[['Attrition_Flag']]).ravel())
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap



from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB, CategoricalNB, BernoulliNB
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import plot_roc_curve

classifiers = [
    SVC(C=0.025),
    DecisionTreeClassifier(max_depth=5),
    RandomForestClassifier(max_depth=5),
     AdaBoostClassifier(),
    GaussianNB(),
    BernoulliNB(),
    LogisticRegression()]
```

```python
fig, ax = plt.subplots()

x_train_preprocessed = preprocessor.fit_transform(X_train)
x_test_preprocessed = preprocessor.transform(X_test)
model_displays = {}
for clf in classifiers:
    clf.fit(x_train_preprocessed, y_train)
    name = type(clf).__name__
    model_displays[type(clf).__name__] = plot_roc_curve(
        clf, x_test_preprocessed, y_test, ax=ax, name=name)
_ = ax.set_title('ROC curve')
```

```python
potential_model = ['SVC', 'DecisionTreeClassifier', 'RandomForestClassifier', 'AdaBoostClassifier', 'LogisticRegression']
for clf in classifiers:
    name = type(clf).__name__
    if name in potential_model:
        predictions =clf.predict(x_test_preprocessed)
        cm = confusion_matrix(y_test, predictions, labels=clf.classes_)
        cr = classification_report(y_true=y_test, 
                                   y_pred=predictions,
                                   target_names=['Existing Customer', 'Attrited Customer'])
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)
        print(f"\n{name}:\n{cr}")
        print(f"{cm}")
        
```
